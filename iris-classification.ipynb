**Stage4: Model Trainer**

```python
#Train_test_split
Train,Test=train_test_split(new_iris_data,test_size=0.2,random_state=42)
```

```python
Train
```

```python
Test
```

```python
Train_x=Train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
Train_y=Train[['Species']]
```

```python
Train_x
```

```python
Train_y
```

```python
Test_x=Test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
Test_y=Test[['Species']]
```

```python
Test_x
```

```python
Test_y
```

```python
#SVM
model=svm.SVC()
```

```python
#training model
model.fit(Train_x,Train_y)
pred=model.predict(Test_x)
```

```python
pred
```

```python
metrics.accuracy_score(pred,Test_y)
```

**Accuracy is 93% with svm**

```python
#Decision Tree
model=DecisionTreeClassifier()
```

```python
#training model
model.fit(Train_x,Train_y)
pred=model.predict(Test_x)
```

```python
metrics.accuracy_score(pred,Test_y)
```

**Accuracy is 93% with Decision Tree**
